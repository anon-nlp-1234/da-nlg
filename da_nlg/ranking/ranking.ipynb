```python

from bleurt import score as blscore

checkpoint = "/Users/apple/Desktop/ml/ucsc/gsr/data2text-nlg/eval/BLEURT/models/BLEURT-20-D3"
scorer = blscore.BleurtScorer(checkpoint)

bleurts = []
for col in (['ref1','ref2','ref3']):
    references = df[col].astype('str').tolist()
    candidates = df['text'].astype('str').tolist()
    scores = scorer.score(references=references, candidates=candidates)
    scores = [round(i,2) for i in scores]
    bleurts.append(scores)
    print(len(scores))

pd.DataFrame(list(zip(bleurts[0],bleurts[1],bleurts[2])),columns=['bleurt_ref1','bleurt_ref2','bleurt_ref3']).to_csv('../data/video_game/temp.csv',index=False)
del scorer
```

```python

from bert_score import score as brscore
berts = []
for col in ['ref1','ref2','ref3']:
    references = df[col].astype('str').tolist()
    candidates = df['text'].astype('str').tolist()
    _,_,scores = brscore(candidates, references, lang="en", verbose=True, rescale_with_baseline=True)
    scores = [round(i,2) for i in scores.tolist()]
    berts.append(scores)
    print(len(scores))

pd.DataFrame(list(zip(berts[0],berts[1],berts[2])),columns=['bertsc_ref1','bertsc_ref2','bertsc_ref3']).to_csv('../data/video_game/temp.csv',index=False)
```

```python
from nltk.translate.meteor_score import meteor_score

meteors = []
for col in ['ref1','ref2','ref3']:
    references = df[col].astype('str').tolist()
    candidates = df['text'].astype('str').tolist()
    # scores = [meteor.compute(predictions=[i], references=[j])['meteor'] for i,j in zip(candidates,references)]
    scores = [meteor_score([j.split()],i.split()) for i,j in zip(candidates,references)]

    scores = [round(i,2) for i in scores]
    meteors.append(scores)
    print(len(scores))

pd.DataFrame(list(zip(meteors[0],meteors[1],meteors[2])),columns=['meteors_ref1','meteors_ref2','meteors_ref3']).to_csv('../data/video_game/temp.csv',index=False)
```

```python
from transformers import AutoModelWithLMHead, AutoTokenizer
import torch
import math
from tqdm import tqdm

gpt2_model = AutoModelWithLMHead.from_pretrained('gpt2')
gpt2_tokenizer = AutoTokenizer.from_pretrained('gpt2')

def score(sentence, tokenizer, model):
    tokenize_input = tokenizer.tokenize(sentence)
    tensor_input = torch.tensor([tokenizer.convert_tokens_to_ids(tokenize_input)])
    out=model(tensor_input, labels=tensor_input)
    return math.exp(out.loss.item())

candidates = df['text'].astype('str').tolist()
scores = [score(i, gpt2_tokenizer, gpt2_model) for i in tqdm(candidates)]

scores = [round(i,2) for i in scores]
print(len(scores))

pd.DataFrame(scores,columns=['PPL']).to_csv('../data/video_game/temp.csv',index=False)
```
